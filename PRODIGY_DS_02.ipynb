```python
import pandas as pd



# Step 1: Find the median age
median_age = data['Age'].median()

# Step 2: Fill missing values with the median
data['Age'] = data['Age'].fillna(median_age)
print(data['Age'])
# Now the 'Age' column has missing values replaced with the median age
```

```python
import pandas as pd

# Assuming your DataFrame is called 'data'

# Step 1: Calculate missing value percentages
missing_value_percentages = (data.isnull().sum() / len(data)) * 100

# Step 2: Find columns with high missing values
columns_with_high_missing = missing_value_percentages[missing_value_percentages > 50].index

# Step 3: Print column names (if any)
if any(missing_value_percentages > 50):
    print("Columns with more than 50% missing values:")
    print(columns_with_high_missing.tolist())
else:
    print("No columns have more than 50% missing values.")

# Step 4: Filter DataFrame (optional, improved)
if any(missing_value_percentages > 50):
    low_missing_columns = missing_value_percentages[missing_value_percentages <= 50].index  # Get column names
    df_filtered = data[low_missing_columns]  # Filter based on list of column names
```

```python
#filling cabin column using simple imputer
from sklearn.impute import SimpleImputer
import pandas as pd


# Step 1: Create a SimpleImputer object
imputer = SimpleImputer(strategy='most_frequent')  # Fills with most frequent value

# Step 2: Fit the imputer on the Cabin column (optional, but recommended)
imputer.fit(data[['Cabin']])  # Learn the most frequent value

# Step 3: Transform (fill) the Cabin column
data['Cabin'] = imputer.transform(data[['Cabin']])
print(data['Cabin'])
# Now the Cabin column has missing values replaced with the most frequent value
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Step 1: Group data by survival (0 or 1) and calculate passenger counts
survival_counts = data.groupby('Survived')['PassengerId'].count()

# Step 2: Create the bar chart
plt.figure(figsize=(8, 6))  # Set figure size
survival_counts.plot(kind='bar', color=['red', 'green'])  # Colors for survived/not survived
plt.title("Survival Rates on the Titanic")
plt.xlabel("Survived (0: No, 1: Yes)")
plt.ylabel("Number of Passengers")
plt.xticks([0, 1], ['No', 'Yes'])  # Set custom x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines

# Optional: Show percentages on top of bars
for i, v in enumerate(survival_counts):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Group data by sex and calculate survival counts (for both survived and not survived)
survival_by_sex = data.groupby('Sex')['Survived'].value_counts().unstack()

# Create the bar chart
plt.figure(figsize=(8, 6))  # Set figure size
survival_by_sex.plot(kind='bar', stacked=False, color=['red', 'green'])  # Colors for survived/not survived

# Labels, title, and grid
plt.xlabel("Sex")
plt.ylabel("Number of Passengers")
plt.title("Survival Rates by Sex on the Titanic")
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: Show percentages on top of bars (modify as needed)
for col in survival_by_sex.columns:
    for i, v in enumerate(survival_by_sex[col]):
        total = survival_by_sex.sum(axis=1)[i]  # Get total passengers per sex group
        percentage = (v / total) * 100  # Calculate percentage
        plt.text(i, v + 0.1, f"{v} ({percentage:.1f}%)", ha='center', va='bottom', fontsize=12)  # Display count and percentage

plt.legend(title='Survived')  # Add legend

plt.show()
```
